```python


#!pip install https://github.com/pandas-profiling/pandas-profiling/archive/master.zip 
from google.colab import drive
drive.mount('/content/drive')
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


from pandas_profiling import ProfileReport


#PublicAssistanceFundedProjectsSummaries
#PublicAssistanceFundedProjectsSummaries
#PublicAssistanceFundedProjectsSummaries
#PublicAssistanceFundedProjectsSummaries
#PublicAssistanceFundedProjectsSummaries

#keep disasterNumber, county, federal obligated amount, incident type, delcarationdate, app name, state

counties = ["Washington Parish", "Orleans Parish", "St. Bernard Parish", "St. Charles Parish", "Jefferson Parish", "Plaquemines Parish", "St. James Parish", "St. John the Baptist Parish", "St. Tammany Parish", "Tangipahoa Parish", "Ascension Parish", "East Baton Rouge Parish", "East Feliciana Parish", "Iberville Parish", "Livingston Parish", "Pointe Coupee Parish", "St. Helena Parish", "West Baton Rouge Parish", "West Feliciana Parish", "Caddo Parish"]
df = pd.read_csv("/content/drive/MyDrive/Team 55 [NEW]/Data/FEMA_cleaned/PublicAssistanceFundedProjectsSummarieskatOnly.csv")

clean_county= pd.Series(counties).str.replace(" Parish", "").tolist()


#keep relevant parishes
df = df[df["county"].isin(clean_county)]

#keep relevant columns for analysis
keep = ["disasterNumber","declarationDate","incidentType", "state","county","applicantName","federalObligatedAmount"]
df = df[keep]

df["year"] = pd.to_datetime(df['declarationDate']).dt.to_period('Y')


######### EDA BELOW --------------------------------------~~~~~~~~~~~~~~~~~~~~

#1d graph of federalObligatedAmount
ax = sns.histplot(data= df['federalObligatedAmount'], log_scale = True, bins = 100)
plt.title("PublicAssistanceFundedProjectsSummaries federalObligatedAmount distribution")
plt.show()


mean = df["federalObligatedAmount"].mean()
sum = df["federalObligatedAmount"].sum()
median = df["federalObligatedAmount"].median()

print("sum:", sum, ",mean:", mean,  ",median:", median)
print()

#% nulls value in each of the columns
size = len(df)
null_count = df.isnull().sum()
null_count = (null_count/float(size))*100
print("% null values")
print(null_count.round(7))
print()


print("years" , list(df["year"].unique()))
print()


#pivot table showing the total aid by county
pt = df.pivot_table('federalObligatedAmount',  columns = "county", aggfunc = "sum")
plt.figure(figsize=(28,10))

plt.bar(pt.columns.tolist(), pt.values[0].tolist())

plt.show()





```

```python

#IndividualsAndHouseholdsProgramValidRegistrations_katOnly.csv
#IndividualsAndHouseholdsProgramValidRegistrations_katOnly.csv
#IndividualsAndHouseholdsProgramValidRegistrations_katOnly.csv
#IndividualsAndHouseholdsProgramValidRegistrations_katOnly.csv
#IndividualsAndHouseholdsProgramValidRegistrations_katOnly.csv
#IndividualsAndHouseholdsProgramValidRegistrations_katOnly.csv

counties = ["Washington Parish", "Orleans Parish", "St. Bernard Parish", "St. Charles Parish", "Jefferson Parish", "Plaquemines Parish", "St. James Parish", "St. John the Baptist Parish", "St. Tammany Parish", "Tangipahoa Parish", "Ascension Parish", "East Baton Rouge Parish", "East Feliciana Parish", "Iberville Parish", "Livingston Parish", "Pointe Coupee Parish", "St. Helena Parish", "West Baton Rouge Parish", "West Feliciana Parish", "Caddo Parish"]
df = pd.read_csv("/content/drive/MyDrive/Team 55 [NEW]/Data/FEMA_cleaned/IndividualsAndHouseholdsProgramValidRegistrations_katOnly.csv")



clean_county= pd.Series(counties).str.replace(" Parish", "").tolist()


df["county"] = df["county"].str.replace(" (Parish)","",regex = False)
df["county"] = df["county"].str.replace(" (County)","",regex = False)

#keep relevant parishes
df = df[df["county"].isin(clean_county)]

keep = ['incidentType', 'declarationDate', 'disasterNumber', 'county','damagedStateAbbreviation',  'grossIncome', 'ownRent', 'primaryResidence','residenceType', 'homeOwnersInsurance', 'floodInsurance','registrationMethod', 'ihpAmount','fipAmount',   'haAmount', 'onaAmount', 'utilitiesOut', 'homeDamage','autoDamage', 'emergencyNeeds', 'foodNeed', 'shelterNeed','accessFunctionalNeeds',   'sbaApproved', 'inspnIssued','inspnReturned', 'habitabilityRepairsRequired', 'rpfvl', 'ppfvl','renterDamageLevel', 'destroyed','floodDamageAmount', 'foundationDamageAmount', 'roofDamageAmount','tsaEligible', 'tsaCheckedIn', 'rentalAssistanceAmount',  'repairAmount','replacementAmount','personalPropertyAmount']


df = df[keep]

df["year"] = pd.to_datetime(df['declarationDate']).dt.to_period('Y')

df["ihp_damage_ratio"] = df["ihpAmount"] / (df["floodDamageAmount"] + df["foundationDamageAmount"] + df["roofDamageAmount"])

test = df["ihp_damage_ratio"].replace([np.inf, -np.inf], np.nan)

test = test.dropna()

#looking at ratio of fema aid to home damage
print("count flood damage",test.min,test.max )


######### EDA BELOW --------------------------------------~~~~~~~~~~~~~~~~~~~~


#get columns that contain amount
#df.drop("")


amount_cols = df.columns[df.columns.str.contains("Amount", regex = False)]
amount_df  = df[amount_cols]

amount_df["county"] = df["county"]

#calculate total amount
amount_df['totalAmount'] = amount_df[amount_cols].sum(axis=1)


zero_count = (amount_df['totalAmount'] == 0).sum().sum()


#drop any 0 values to not interfere with histplot
amount_df = amount_df.replace(0,np.nan)


ax = sns.histplot(data=amount_df["totalAmount"], log_scale = True, bins = 30)
plt.title("IndividualsAndHouseholdsProgramValidRegistrations total aid distribution")
plt.show()


mean = amount_df["totalAmount"].mean()
sum = amount_df["totalAmount"].sum()
median = amount_df["totalAmount"].median()
min = amount_df["totalAmount"].min()
max = amount_df["totalAmount"].max()

print("sum:", sum, ",mean:", mean,  ",median:", median, ",min:" , min , ",max:" , max)
print()


print("zero_count: ", zero_count)
print()

#% nulls value in each of the columns
size = len(df)
null_count = df.isnull().sum()
null_count = (null_count/float(size))*100
print("% null values")
print(null_count)
print()

print("years" , list(df["year"].unique()))


#pivot table showing the total aid by county
pt = amount_df.pivot_table('totalAmount',  columns = "county", aggfunc = "sum")


plt.figure(figsize=(28,10))

plt.bar(pt.columns.tolist(), pt.values[0].tolist())

plt.show()




```
