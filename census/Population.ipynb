```python
# 2000-2020 Population count
# Source:  U.S. Census Bureau
# Release date: 

    
earlyFile= '/content/drive/MyDrive/Team 55 [NEW]/Data/census/Population/co-est00int-01-22.xlsx'
lateFile = '/content/drive/MyDrive/Team 55 [NEW]/Data/census/Population/co-est2020.csv'

upTo2010 = pd.read_excel(earlyFile, sheet_name= 'CO-EST00INT-01-22', header=3 , skipfooter = 8).fillna(method='ffill')
post2010 = pd.read_csv(lateFile,engine='python',encoding='latin1')

upToCol = {'Unnamed: 1':'CENSUS2000POP','Unnamed: 0':'CTYNAME','Unnamed: 12': 'CENSUS2010POP', 'Unnamed: 13':2010}
upTo2010.rename(columns=upToCol ,inplace=True)
postCol = {'POPESTIMATE2010':2010,
       'POPESTIMATE2011':2011, 'POPESTIMATE2012':2012, 'POPESTIMATE2013':2013,
       'POPESTIMATE2014':2014, 'POPESTIMATE2015':2015, 'POPESTIMATE2016':2016,
       'POPESTIMATE2017':2017, 'POPESTIMATE2018':2018, 'POPESTIMATE2019':2019,
       'POPESTIMATE042020':'CENSUS2020POP', 'POPESTIMATE2020':2020}
post2010.rename(columns=postCol,inplace=True)

#isolating louisianas data only
post2010 = post2010.loc[post2010['STATE'].astype(int) == 22]

#removing columns we don't need, like state and statenumbers since everything is now louisiana data
post2010 = post2010.drop(post2010.columns[[0, 1, 2,3,4,5,7,8,9]],axis = 1)
#city names had . before each one, remove it
upTo2010['CTYNAME'] = upTo2010['CTYNAME'].str[1:]

louisianaCleanPop = pd.merge(upTo2010, post2010, on='CTYNAME')

louisianaCleanPop.head()
```
