```python

#single family homes time series
df_list = []

tiers =["_0.0_0.33_sm_sa_month.csv" ,"_0.33_0.67_sm_sa_month.csv","_0.67_1.0_sm_sa_month.csv" ]

for i in range(3):
  df = pd.read_csv("/content/drive/MyDrive/Team 55 [NEW]/Data/zillow_data_update/county/County_zhvi_uc_sfr_tier" + tiers[i] )
  df = df[df["State"] == "LA"]
  df = df.melt(id_vars=["RegionID",	"SizeRank",	"RegionName",	"RegionType",	"StateName",	"State",	"Metro",	"StateCodeFIPS",	"MunicipalCodeFIPS"], var_name="Date", value_name="Value")
  if i == 0:
    df["tier"] = "0.00_0.33"

  if i == 1:
    df["tier"] = "0.33_0.67"
  
  if i == 2:
    df["tier"] = "0.67_1.0"
  df_list.append(df)
  del(df)
  gc.collect()
  df=pd.DataFrame()

merged_df = pd.concat(df_list)

merged_df.to_csv('/content/drive/MyDrive/Team 55 [NEW]/Data/zillow_data_update/zillow_single_family_homes.csv', index=False)  


```
